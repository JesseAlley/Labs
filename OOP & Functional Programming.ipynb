```python
class RecursiveSearch:
    def binarysearch(self,array, low, high, key):

        if low <= high:
            mid = low + (high - low) // 2

            if array[mid] == key:
                return mid

            elif array[mid] < key:
                return self.binarysearch(array, mid+1, high, key)

            else:
                return self.binarysearch(array, low, mid-1, key)
        else:
            return -1

array = [1,2,3,4,5]
key = 4

solution = RecursiveSearch()
target_index = solution.binarysearch(array, 0, len(array)-1, key)

if target_index != -1:
    print(f"Target found at index {target_index}")
else:
    print("Target not found in array")
```
